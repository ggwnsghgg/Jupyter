```python
# CGV 영화 예매
# 일반, 학생에 따라서 영화 요금 출력 프로그래밍 

# 함수명 정리

# 영화 = movie
# 영화표 가격 = money
# 2D영화 = 2d
# 3D영화 = 3d
# 좌석 위치 프리미엄 = prime
# 좌석 위치 이코노미 = economy
# 학생, 성인 = age
# 요일 = day
```

```python

# print("안녕하세요 CGV입니다")
# print("영화를 선택해주세요")
day = ["월요일","월","화요일","화","수요일","수","목요일","목","금요일","금"]
day2 = ["일요일","일","토요일","토"]
money = 7000


####################################################  

        #나이 input 구문 (나이 값 받아오는 부분)


while True:                                                           #예외처리 하는 부분
    try:
        age = int(input("나이를 입력해주세요 : "))
        break
    except ValueError:
        print("나이를 다시 입력해주세요")

        
####################################################    
    

    # 나이체크 하는 부분
    
    
if age <= 19:
    
    
    # 학생 요금
    
    print("학생입니다.")
    mon = money - 2000  # 학생 할인요금
    
    
####################################################  
    
    
    # 상영관 요금제

    while True:
        try:
            movie = input("2D 와 3D관 관람하실 관을 선택해주세요 : ")
            if "2D" in movie:
                print("2D관을 선택하셨습니다.")
                mon2 = mon + 3000
                break
            
            elif "3D" in movie:
                print("3D관을 선택하셨습니다.")
                mon2 = mon + 5000
                break
            
            else:
                print("오류입니다.")
  
        except ValueError:
            print("상영관을 다시 선택해주세요")

            
    
####################################################  
    
    
    
    # 평일 , 주말 요금제
    while True:
        try:    
            dayinput = input("요일을 입력해주세요 : ")
            if dayinput in day2:
                print("주말입니다.")
                mon3 = mon2 + 2000   #주말 추가요금 
                print("가격은 " + str(mon3) + " 입니다. ")
                break
            
            elif dayinput in day:
                print("평일입니다.")
                print("가격은 " + str(mon2) + " 입니다. ")
                break
            
            else :
                print("오류입니다")
            
        except ValueError:
                print("요일을 다시 입력해주세요")
    
else:                                     
    
    

###################################################### 



    # 나이 체크 하는 부분 

    
    if age >= 20:   
        print("성인입니다.")
        mon = money + 2000 
        
        
        
        
######################################################
    
    
    
    
    # 상영관 요금제
        while True:
            try:
                movie = input("2D 와 3D관 관람하실 관을 선택해주세요 : ")
                if "2D" in movie:
                    print("2D관을 선택하셨습니다.")
                    mon2 = mon + 3000
                    break
            
                elif "3D" in movie:
                    print("3D관을 선택하셨습니다.")
                    mon2 = mon + 5000
                    break
            
                else:
                    print("오류입니다.")
 
            except ValueError:
                    print("상영관을 다시 선택해주세요")

    
    
    
    
######################################################
    
    
    
    
    
    # 평일 , 주말 요금제
        while True:
            try:    
                dayinput = input("요일을 입력해주세요 : ")
                if dayinput in day2:
                    print("주말입니다.")
                    mon3 = mon2 + 2000   #주말 추가요금 
                    print("가격은 " + str(mon3) + " 입니다.")
                    break
                
                elif dayinput in day:
                    print("평일입니다.")
                    print("가격은 " + str(mon2) + " 입니다. ")
                    break
                else :
                    print("오류입니다")

            except ValueError:
                print("요일을 다시 입력해주세요")

```
